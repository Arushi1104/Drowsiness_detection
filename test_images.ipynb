```python
import cv2
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Load the saved model
model = tf.keras.models.load_model('models/drowsy.h5')

# Read the image
img = cv2.imread('Screenshot 2025-03-24 164057.png')

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Resize the image to match model input size
img_resized = tf.image.resize(img_rgb, (256, 256))

# Normalize the image (assuming the model expects values between 0 and 1)
img_normalized = img_resized / 255.0

# Expand dimensions to match model input shape
img_input = np.expand_dims(img_normalized, axis=0)

# Make prediction
yhat = model.predict(img_input)

# Display the image
plt.imshow(img_rgb)
plt.axis('off')
plt.show()

# Print prediction result
predicted_class = 'awake' if yhat > 0.5 else 'drowsy'
print(f'Predicted class is {predicted_class}')
```
